```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x,y)
plt.show()
```

```python
#first method 
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
# second method
fig=plt.figure()
ax=fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

```python
# 3rd method and recommended
fig,ax=plt.subplots()
ax.plot(x,y);
```

## Matplotlib example workflow

```python
# Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#setup the plot
fig,ax = plt.subplots(figsize=(10,10)) #(width,height)

# plot data
ax.plot(x,y)

# customize the plot
ax.set(title="Simple plot",xlabel="x-axis" , ylabel="y-axis")

# fig.savefig("path") # this saves the image of the graph in the given path
```
