## Matplotlib example workflow

```python
# Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#setup the plot
fig,ax = plt.subplots(figsize=(10,10)) #(width,height)

# plot data
ax.plot(x,y)

# customize the plot
ax.set(title="Simple plot",xlabel="x-axis" , ylabel="y-axis")

# fig.savefig("path") # this saves the image of the graph in the given path
```

## Two options for subplot

```python
#option 1 for subplot

fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2, figsize=(10,6))

ax1.plot(x,x**2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.hist(y);
ax4.hist(z);
```

```python
#option 2 for subplots

fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,6))
ax[0,0].plot(x,x**2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].hist(y);
ax[1,1].hist(z);
```

```python
# read data from heart disease csv
heart_disease= pd.read_csv("087 heart-disease.csv")
heart_disease.head()
# len(heart_disease)
```

```python
heart_disease["age"].plot.hist(bins=30)
```

```python
heart_disease.plot.hist(figsize=(20,40) ,subplots=True)
```

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
## pyplot method
over_50.plot(kind="scatter",x="age",y="chol",c="target")
```

```python
# using the OO method mixed with pyplot
fig,ax=plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
                    x="age",
                    y="chol",
                    c="target",
                    ax=ax)
ax.set_xlim([45,100])
```

```python
## using OO method from scratch
fig,ax = plt.subplots(figsize=(10,6))

# plot the data
scatter=ax.scatter(over_50["age"],
                   over_50["chol"],
                  c=over_50["target"]);
# customization
ax.set(title="cholestrol level",
        xlabel="age",
        ylabel="chol");

# add a legend
ax.legend(*scatter.legend_elements(),title="Target");

# add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--")
```

```python
# plotting cholestrol levels

fig, (ax0,ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10));

ax0.scatter(x=heart_disease["age"],
            y=heart_disease["chol"],
            c=heart_disease["target"])

ax0.set(title="cholestrol level",
        xlabel="age",
        ylabel="chol")

ax0.legend(*scatter.legend_elements());

ax0.axhline(heart_disease["chol"].mean(),
           linestyle="--")


# plotting thalach 

ax1.scatter(x=heart_disease["age"],
            y=heart_disease["thalach"],
            c=heart_disease["target"])

ax1.set(title="thalach level",
        xlabel="age",
        ylabel="thalach")

ax1.legend(*scatter.legend_elements());

ax1.axhline(heart_disease["thalach"].mean(),
           linestyle="--")

fig.suptitle("Heart disease analysis",fontsize=18,fontweight="bold");
```

## Customizing using matplotlib

```python
# changing the style 
plt.style.use("seaborn-v0_8-whitegrid")

## using OO method from scratch
fig,ax = plt.subplots(figsize=(10,6))

# plot the data
scatter=ax.scatter(x=over_50["age"],
                   y=over_50["chol"],
                  c=over_50["target"],
                  cmap="winter");
# customization
ax.set(title="cholestrol level",
        xlabel="age",
        ylabel="chol");

# add a legend
ax.legend(*scatter.legend_elements(),title="Target");

# add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--")
```

```python
# customizing the x axis and y axis limitations
# plotting cholestrol levels

fig, (ax0,ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10));

ax0.scatter(x=heart_disease["age"],
            y=heart_disease["chol"],
            c=heart_disease["target"],
           cmap="winter")

#ax0.set_xlim([30,80])

ax0.set(title="cholestrol level",
        xlabel="age",
        ylabel="chol")

ax0.legend(*scatter.legend_elements());

ax0.axhline(heart_disease["chol"].mean(),
           linestyle="--")


# plotting thalach 

ax1.scatter(x=heart_disease["age"],
            y=heart_disease["thalach"],
            c=heart_disease["target"],
           cmap="winter")

ax1.set(title="thalach level",
        xlabel="age",
        ylabel="thalach")
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
ax1.legend(*scatter.legend_elements());

ax1.axhline(heart_disease["thalach"].mean(),
           linestyle="--")

fig.suptitle("Heart disease analysis",fontsize=18,fontweight="bold");
```
