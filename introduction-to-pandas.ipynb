## Making new columns

```python
#columns from series
car_seats= pd.Series([5,5,5,5,5])

# adding this seats columns to car_sales
car_sales["Seats"] = car_seats

car_sales
```

```python
# some data have NaN. we will replace that
car_sales.Seats = car_sales.Seats.fillna(5)

car_sales
```

```python
# making columns from a python list
fuel_eco = [7,9,33,8,1,2,3,4,5,6]

car_sales["fuel per 100km"] = fuel_eco
car_sales
```

```python
car_sales["Total fuel used(L)"] = car_sales["fuel per 100km"]/100 * car_sales["Odometer (KM)"]
car_sales
```

```python
car_sales["Wheels"] = 4
car_sales
```

```python
car_sales.dtypes
```

```python
car_sales=car_sales.drop("Wheels" ,axis=1) # remove a column using axis=1(column)
car_sales
```

```python
# select a random sample
car_sales_shuffeled = car_sales.sample(frac=1)
car_sales_shuffeled
```

```python
car_sales_shuffeled.reset_index(drop=True) # this is used to reset the suffled index. the drop=True ensures that the old indexes the not repeated as a column
```

```python
# converting the km to miles 
car_sales["Odometer (KM)"]= car_sales["Odometer (KM)"].apply(lambda x : x/1.6)
car_sales
```
