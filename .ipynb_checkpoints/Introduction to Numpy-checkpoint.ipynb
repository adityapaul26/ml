## Aggregation

Aggregation = performing the same operations of multiple items

```python
sum(a1)
```

```python
np.sum(a1)
```

```python
#use python functions for python datatypes
#use numpy func for numpy datatypes
```

```python
massive_array= np.random.random(1000000)
massive_array
```

```python
%timeit sum(massive_array)
%timeit np.sum(massive_array)
```

```python
# np.sum is massively faster than sum func of python almost 526 times faster
```
