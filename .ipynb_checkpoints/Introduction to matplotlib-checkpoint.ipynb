# Introduction to matlplotlib

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.plot();
```

```python
plt.plot()
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
plt.plot(x,y)
plt.show()
```

```python
#first method 
fig = plt.figure()
ax = fig.add_subplot()
plt.show()
```

```python
# second method
fig=plt.figure()
ax=fig.add_axes([1,1,1,1])
ax.plot(x,y)
plt.show()
```

```python
# 3rd method and recommended
fig,ax=plt.subplots()
ax.plot(x,y);
```

## Matplotlib example workflow

```python
# Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

#setup the plot
fig,ax = plt.subplots(figsize=(10,10)) #(width,height)

# plot data
ax.plot(x,y)

# customize the plot
ax.set(title="Simple plot",xlabel="x-axis" , ylabel="y-axis")

# fig.savefig("path") # this saves the image of the graph in the given path
```

## Making figures with numpy arrays

```python
x = np.linspace(0,10,100)
x
```

### Line 

```python
fig,ax=plt.subplots()
ax.plot(x,x**2);
```

### Scatter plot

```python
# use same data to create a scatter plot
fig,ax=plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
fig,ax=plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
# makeing a scatter plot from a dictionary
z = {1:1,2:4,3:6,4:8}
fig,ax = plt.subplots()
ax.bar(z.keys(),z.values());
ax.set(title="Doubles",ylabel="twice",xlabel="nums")
```

```python
fig,ax = plt.subplots()
ax.barh(z.keys(),z.values());
```

```python
fig,ax = plt.subplots()
ax.hist(z.keys(),z.values());
```

### histogram

```python
y = np.random.randn(1000) # returns a sample from standard normal distribution
fig,ax= plt.subplots()
ax.hist(y)
```

## Two options for subplot

```python
#option 1 for subplot

fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2, figsize=(10,6))

ax1.plot(x,x**2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.hist(y);
ax4.hist(z);
```

```python
#option 2 for subplots

fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10,6))
ax[0,0].plot(x,x**2);
ax[0,1].scatter(np.random.random(10),np.random.random(10));
ax[1,0].hist(y);
ax[1,1].hist(z);
```

```python
car_sales["Price"]= car_sales["Price"].str.replace('[\$\,\.]','',regex=True) #in newer version of python regex is set to false .. so we need to make it true
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].str[:-2]
car_sales
```

```python
car_sales["sale date"]=pd.date_range('1/1/2026',periods=len(car_sales))
car_sales
```

```python
car_sales["Total sales"]=car_sales["Price"].astype(int).cumsum()
car_sales
```

```python
plt.plot(car_sales["sale date"],car_sales["Total sales"]);
#or 
car_sales.plot(x="sale date",y="Total sales")
```
